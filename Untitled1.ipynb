```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
import matplotlib.pyplot as plt
```

```python
img=imread("C:/Users/admin/Pictures/cv/butterfly1.jfif")
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
c=resize(img,(128*10,64*10))
plt.axis("off")
plt.imshow(c)
plt.show()
print(c.shape)
```

```python
fd,hog_image = hog(img,orientations=10,pixels_per_cell=(2,2),cells_per_block=(4,4),visualize=True,channel_axis=2)
print(fd.shape)
print(hog_image.shape)
plt.axis("off")
plt.imshow(hog_image,cmap="hot")
plt.show()
```

```python
fd,hog_image = hog(img,orientations=10,pixels_per_cell=(2,2),cells_per_block=(4,4),visualize=True,channel_axis=2)
print(fd.shape)
print(hog_image.shape)
plt.axis("off")
plt.imshow(hog_image,cmap="ocean_r")
plt.show()
```
